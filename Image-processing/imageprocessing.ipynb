```python
import numpy as np
import cv2
import matplotlib.pyplot as plt

# Load the image in grayscale mode 
img = cv2.imread("Images/sample.jpg") # ,cv2.IMREAD_GRAYSCALE for grayscale

# Display the image 
# plt.imshow(img, cmap='gray')
# plt.title("Grayscale Image")
# plt.axis("off")
# plt.show()
```

```python
# Convert image to grayscale format
gray_image = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
plt.imshow(gray_image, cmap='gray')
plt.title("Grayscale Image")
plt.show()
```

```python
# Flipping an image
flipped = cv2.flip(img, -1)  # 1 = Horizontal, 0 = Vertical, -1 = Both
plt.imshow(cv2.cvtColor(flipped, cv2.COLOR_BGR2RGB))
plt.title("Flipped Image")
plt.show()
```

```python
# Edge detection using Sobel filter 
edges = cv2.Sobel(img, cv2.CV_64F, 1, 1, ksize=5)
plt.imshow(edges, cmap='gray')
plt.title("Edge Detection")
plt.show()
```

### Apply the Gassian Blur to smooth out the image 

```python
image = cv2.imread("Images/sample.jpg")

if image is None: 
    print("Error: Could not open or read the image file.")
else:
    
    # Convert BGR to RGB for correct color display in Matplotlib
    image_rgb = cv2.cvtColor(image, cv2.COLOR_BGR2RGB)

    # Apply Gaussian blur
    blurred_image = cv2.GaussianBlur(image,(15,15),50)

    # Display the original and blurred images
    plt.subplot(1, 2, 1)
    plt.imshow(image_rgb)
    plt.title("Original Image")
    
    plt.subplot(1, 2, 2)
    plt.imshow(cv2.cvtColor(blurred_image, cv2.COLOR_BGR2RGB))  # Convert back to RGB
    plt.title("Blurred Image (Gaussian Blur)")
    
    plt.show()
```

```python
# Another example to visualize the diffrence 
# Load the image
image1 = cv2.imread("Images/sample.jpg")

# Apply Gaussian Blur with different sigma values
blur_0 = cv2.GaussianBlur(image1, (15, 15), 0)   # Automatic sigma
blur_50 = cv2.GaussianBlur(image1, (15, 15), 50) # Manually set sigma

# Convert images to RGB for correct display
image_rgb = cv2.cvtColor(image1, cv2.COLOR_BGR2RGB)
blur_0_rgb = cv2.cvtColor(blur_0, cv2.COLOR_BGR2RGB)
blur_50_rgb = cv2.cvtColor(blur_50, cv2.COLOR_BGR2RGB)

# Display the images
plt.figure(figsize=(10, 5))

plt.subplot(1, 3, 1)
plt.imshow(image_rgb)
plt.title("Original Image")

plt.subplot(1, 3, 2)
plt.imshow(blur_0_rgb)
plt.title("Blurred (sigma = 0)")

plt.subplot(1, 3, 3)
plt.imshow(blur_50_rgb)
plt.title("Blurred (sigma = 50)")

plt.show()
```

## Crop a specific part of the image

```python
# crop (y_start:y_end, x_start:x_end)
print("The shape of image is: ",image1.shape) # shape(height,width,channels)
if image1 is None:
    print("Error: Could not open or read the image file.")
else:
    cropped_image = image1[500:2000, 800:3200]

    cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)

    #Display the cropped image
    plt.imshow(cropped_image_rgb)
    plt.title("Cropped Image")
    plt.axis("off")
    plt.show()
```

```python
# Alternatively center crop 
height, width, _ = image1.shape

# Define center crop size (e.g., 50% of the image)
crop_h, crop_w = height // 2, width // 2

# Compute the start and end points
y_start, y_end = (height - crop_h) // 2, (height + crop_h) // 2
x_start, x_end = (width - crop_w) // 2, (width + crop_w) // 2

# Crop the image
cropped_image = image[y_start:y_end, x_start:x_end]

# Convert and display
cropped_image_rgb = cv2.cvtColor(cropped_image, cv2.COLOR_BGR2RGB)
plt.imshow(cropped_image_rgb)
plt.title("Center Cropped Image")
plt.axis("off")
plt.show()
```

### Perfromaing Object detection using Canny edge

```python
# Convert image to grayscale
gray = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)

# Apply Canny Edge Detection
edges = cv2.Canny(gray, 50, 150)

# Display the edge-detected image
plt.imshow(edges, cmap="gray")
plt.title("Edge Detection (Canny)")
plt.show()
```
