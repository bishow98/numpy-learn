```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
sales_data = np.array([
    [1, 150000, 180000, 220000, 250000],  # Paradise Biryani
    [2, 120000, 140000, 160000, 190000],  # Beijing Bites
    [3, 200000, 230000, 260000, 300000],  # Pizza Hub
    [4, 180000, 210000, 240000, 270000],  # Burger Point
    [5, 160000, 185000, 205000, 230000]   # Chai Point
])

print("----------- Foodmandu Sales analysis ---------- ")
print("\n Shape of sales data: ",sales_data.shape )
print("Selecting the first 3 row:\n", sales_data[:3])
print("\n Selecting the row and column: \n", sales_data[:, 1:])
```

```python
# total sales per year 
print(np.sum(sales_data, axis=0))
yearly_total = np.sum(sales_data[:, 1:],axis=0)
print(yearly_total)
total_of_all_year = np.sum(yearly_total)
print("Total sales in 4 years: ",total_of_all_year)


#Minimum sales per restarurats
min_sales = np.min(sales_data[:, 1:], axis=1)
print(min_sales)

#Minimum sales of paradise biryani 
min_sales_paradise = np.min(sales_data[0, 1:])
print("Minimum sales of the paradise biryani is : ",min_sales_paradise)

#Maximum sales of restaurant
max_sales = np.max(sales_data[:, 1:], axis=1)
print("\n Maximum sales of restaurant is: \n",max_sales)

#Maximum sales year 
max_sales_year = np.max(sales_data[:,1:],axis=0)
print("\n Maximum sales based on year is: \n",max_sales_year)

#avg sales per restaurant 
avg_sales = np.mean(sales_data[:, 1:], axis=1)
print("\n Average sales per restaurant is: \n",avg_sales)
```

```python
cumsum = np.cumsum(sales_data[:, 1:],axis=1)
print("Cumulative sum : \n", cumsum)
mean_cumsum = np.mean(cumsum, axis=0)
print(mean_cumsum)

plt.figure(figsize=(10,6))
plt.plot(np.mean(cumsum, axis=0))
plt.title("Average Cumulative sales across all restarurant")
plt.xlabel("Years")
plt.ylabel("sales")
plt.grid(True)
plt.show()
```

```python
vector1 = np.array([1,2,3,4,5])
vector2 = np.array([6,7,8,9,10])

print("Vector addition: ", vector1 + vector2)

print("\n Multiplication vector ", vector1 * vector2)

print(" \n Dot Product ", np.dot(vector1, vector2))

print(" \n Magnitude of vector1: ",np.linalg.norm(vector1))

print(" \n magnitude of vector2 : ",np.sqrt(vector1.dot(vector1)))

angle = np.arccos(np.dot(vector1,vector2) / (np.linalg.norm(vector1)* np.linalg.norm(vector2)))

print("The angle of two vector is : ", angle) 

```
